```python
symbols = ['BTC', 'ETH', 'BNB', 'XRP']
intervals = ["1d", "12h", "6h", "1h"]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_coin(symbol, interval):
    file_path = f"./data/{symbol}_{interval}.csv"
    read_file = pd.read_csv(file_path)

    read_file["open_time"] = pd.to_datetime(read_file["open_time"])

    january_first_indices = [(i, date.year) for i, date in enumerate(read_file["open_time"]) if date.month == 1 and date.day == 1]

    plt.figure(figsize=(10, 6))

    plt.plot(list(range(len(read_file))), np.log10(read_file["open"]), label='Open', color="teal")
    plt.plot(list(range(len(read_file))), np.log10(read_file["close"]), label='Close', color="darkblue")

    for i, year in january_first_indices:
        plt.axvline(x=i, color='gray', linestyle='--', alpha=0.6)
        plt.text(i, plt.ylim()[1] - 0.05, str(year), verticalalignment='top', fontsize=12)

    plt.xlabel("Time")
    plt.ylabel("Price in log10")
    plt.title(f"{symbol} Open and Close Price Over Time in {interval} interval")
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 6))

    plt.plot(list(range(len(read_file))), np.log10(read_file["low"]), label='Low', color="red")
    plt.plot(list(range(len(read_file))), np.log10(read_file["high"]), label='High', color="green")

    for i, year in january_first_indices:
        plt.axvline(x=i, color='gray', linestyle='--', alpha=0.6)
        plt.text(i, plt.ylim()[1] - 0.05, str(year), verticalalignment='top', fontsize=12)

    plt.xlabel("Time")
    plt.ylabel("Price in log10")
    plt.title(f"{symbol} Low and High Price Over Time in {interval} interval")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
for symbol in symbols:
    for interval in intervals:
        if interval == "1d":
            plot_coin(symbol, interval)
```

```python
for i, s in enumerate(symbols):
    df = pd.read_csv(f"./data/{s}_1h.csv")
    plt.plot(np.log10(df["high"]), label=s)
    plt.legend()
```
